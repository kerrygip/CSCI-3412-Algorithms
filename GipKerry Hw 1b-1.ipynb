```python
#Name: Kerry Gip
#Date: 1/31/2021

import random

#PART 1: Binary search simulation  (Power of divide by 2) : (5 points)
     
#https://www.geeksforgeeks.org/python-program-for-binary-search/ making sure I have the binary search correct        
def binarySearch(someList, rando):
    #define parameters
    count = 0
    low = 0
    mid = 0
    high = len(someList)
    found = False
    
    while (found != True):
        mid = (low+high)//2
        
        #checks if rando is in the right array
        if (someList[mid] < rando):
            #makes new low at the halfway point
            low = mid+1
            count = count +1
#            print("Random (", rando, ") is higher than mid. low = ", low, ", high = ", high, " count = ", count)
        #else rando is in the left array
        elif (someList[mid] > rando):
            #high point starts at less than middle
            high = mid-1
            count = count +1
#            print("Random (", rando, ") is lower than mid. low = ", low, ", high = ", high, " count = ", count)
        
        #(someList[mid] == rando)
        else:
            found = True
            
    
    return count
#guess the number loop?
        #count = count/10000


#function call
def countNum(size):
    someList = list(range(size))
    count = 0
    for i in range(size):
    #gen random
        rando = random.randrange(1000)
    # binsearch: countS
        count += binarySearch(someList, rando)
    return count
    

print("Total count for 10k: ", countNum(10000), "Average count: ", countNum(10000)/10000)
print("Total count for 1000:", countNum(1000), "Average count: ", countNum(1000)/10000)
print("Total count for 1 million:", countNum(1000000), "Average count: ", countNum(1000000)/10000)

#results = binarySearch(someList, rando)

#print results
#print("First number is:", rando, "It took: ", results,  "tries to match up")
```

```python
#PART 4: Display top N number of words in descending order of a file

def main():
         
    count = 0
    keepLooping = True
    newDict = dict()
    
    #infile check
    while keepLooping:
        try:
            infileName = input("Enter the file name: ")
            infile= open(infileName, "r")
            keepLooping = False
        except FileNotFoundError:
            print("Invalid file name, try again")
    
    #infile read/strip/split/lower 
    #reads through the infile and adds to dictionary
    for line in infile:
        words = line.strip().lower().split()
        for word in words:
            if word in newDict:
                newDict[word] = newDict[word] + 1
                #if not in dicitonary, assign it  value 1 
            else:
                newDict[word] = 1

    #print results
    print("Words and their frequencies")
    for key in list(newDict.keys()):
        print(key, newDict[key])
main()       
```
