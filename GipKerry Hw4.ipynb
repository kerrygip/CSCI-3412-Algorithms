[Wiki Link on Linear vs Constant Time](https://en.wikipedia.org/wiki/Time_complexity#:~:text=An%20algorithm%20is%20said%20to,the%20size%20of%20the%20input.&text=Hence%20it%20is%20a%20linear,taking%20O(n)%20time.)  

An algorithm is said to be constant time (also written as O(1) time) if the value of T(n) is bounded by a value that does not depend on the size of the input.  
For example, accessing any single element in an array takes constant time as only one operation has to be performed to locate it. In a similar manner, finding the minimal value in an array sorted in ascending order; it is the first element.  

However, finding the minimal value in an unordered array is not a constant time operation as scanning over each element in the array is needed in order to determine the minimal value. Hence it is a linear time operation, taking O(n) time. If the number of elements is known in advance and does not change, however, such an algorithm can still be said to run in constant time.

**Search(L,K)**	- List- Search takes O(n) time in the worst case if it has to seach entire list, for singly or doubly  linked list, unsorted or sorted   

**Insert(L,x)** - The running time for List-Insert on a list of n elements is O(1) for anything unsorted, since it just adds to the front of the list and that takes constant time. It takes O(n) to insert as it has to sort the list first and then add the element in.   

**Delete(L,x)** - List Delete runs in O(1) time since it just deletes the first element, but if we wish to delete an element with a given key(like finding it sorted), then O(n) is required for the worst case since we have to call List-Search to find it first.   

**Successor(L,x)** - Would have to refer to the wiki link where, accessing a single element in the array takes O(1) time since there is only one operation that is needed to be performed on it in an ordered list. For an unsorted list, we would have to sort it O(n) time to find it's suceessor   

**Predecessor(L,x)** - It is O(1) for the sorted doubly linked list because we can just go look at the tail node and look at the previous pointer. It is also instantaneous, since we can just look at what the predecessor pointer is since it is already sorted.  
For a singly linked list, sorted or not, we have to traverse the whole list to find the predecessors. For an unsorted doubly linked list, we would need to sort the list to find the predecessor. All these times take O(n) time. 


**Minimum(L)** - The same rules apply for sorted/unsorted. Sorted lists take O(1) time since we can find the minimum instantly at the front of the list and would need O(n) to sort the list or traverse it to find the min. 

**Maximum(L)** - It would take O(n) no matter what since we would have to traverse the list to find the maximum at the end of the list.

```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
import requests 
import json

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup

#https://www.scrapehero.com/a-beginners-guide-to-web-scraping-part-2-build-a-scraper-for-reddit/

url = "http://www.ucdenver.edu/pages/ucdwelcomepage.aspx"
html = urlopen(url)

#making request instance
requests = requests.get(url, headers = {'User-agent': 'your bot 0.1'})

#using module
soup = BeautifulSoup(requests.content, "html.parser")
#print(soup)

#display title
# print("Title of the website is: ")
# for title in soup.find_all("title"):
#     print(title.get_text())

siteTable = soup.find("div",attrs={'id':'horizontal-nav-collapse'})  

#look at dept, name, url

links = siteTable.find_all("a")    

urls = []
for link in links: 
    title = link.text.strip()  
    url = link['href']
    if title != '':
        if not url.startswith('http'):    
            url = "https://www.ucdenver.edu"+url
        record = {
            'title': title,
            'url':url
        }
        urls.append(record)


with open('data.json', 'w') as outfile:
    json.dump(urls, outfile, sort_keys=True)
    
table = pd.read_json('Ucdenver_link_data_new.json') 
table
```

```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
import requests 
import json
import os 
import lxml.html as html
%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup

#using links
#https://towardsdatascience.com/web-scraping-html-tables-with-python-c9baba21059
#https://www.scrapehero.com/a-beginners-guide-to-web-scraping-part-2-build-a-scraper-for-reddit/


url = "https://covid19.colorado.gov/data"
request = requests.get(url, headers = {'User-agent': 'your bot 0.1'})

page = html.fromstring(request.content)

tr_elements = page.xpath('//table')

#Create empty list
col=[]
i=0
#For each row, store each first element (header) and an empty list
for t in tr_elements[0]:
    i+=1
    name=t.text_content()
    print('%d:"%s"'%(i,name))
    col.append((name,[]))
    
#creating pandas dataframe
#Since out first row is the header, data is stored on the second row onwards
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 10, the //tr data is not from our table 
    if len(T)!=10:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
       
    
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)        
```
