```python
#Kerry Gip
#CSCI 3412 003
#Due 2/21/21


#rand1000.txt
import time

def readData(numList):

    infileName = input("What is the name of the file you wish to enter?: ")  

    try:               
        infile = open(infileName, "r")
        for line in infile:
            for i in line.strip().split():
                numList.append(int(i))

       #Display error message if the input file is not found.     
    except FileNotFoundError:
        print("File not found")
    
    infile.close()
    
def insertionSort(numList):
    
    # https://www.geeksforgeeks.org/python-program-for-insertion-sort/
    # Traverse through 1 to len(numbers) 
    count = 0
    for i in range(1, len(numList)):
        
        #sets current position as sorted and remembers the first number of the list
        current = numList[i] 
        #position/index starts at 0
        position = i-1
        count = count +1
        
        #loops it all the way to the end of the list
        #if current is less than the numList at whatever position it is
        #position moves to the right
        while position >= 0 and current < numList[position] : 
                #swap
                numList[position + 1] = numList[position] 
                position = position -  1
                count = count +1
                #inserts unsorted element to correct position
        numList[position + 1] = current
        
        
    print("The number of insertion comparisons is: ", count)
    
def mergeSort(numList):
        #https://www.geeksforgeeks.org/merge-sort/
         # Traverse through 1 to len(numbers) 
    count = 0
    if len(numList) > 1:
 
         # Finding the middle of the list
        mid = len(numList)//2
 
        # Dividing the list into two halves
        leftList = numList[:mid]
        rightList = numList[mid:]
 
        #splitting the unsorted groups recursively until there is 1 element per group
        leftRecur = mergeSort(leftList)
        rightRecur = mergeSort(rightList)
         
        #also have to count the nmber of comparisons in the recursions   
        count += leftRecur + rightRecur
        
        indexLeft = 0
        indexRight = 0
        indexNumList = 0
        
 
        # while is just making sure the start position is less than length of list 
        while indexLeft < len(leftList) and indexRight < len(rightList):
            count = count +1
            
            #we are sorting the left list first before moving to the right list and moving towards the while loop
            if leftList[indexLeft] < rightList[indexRight]:
                #the original list index is now the index of the left list
                numList[indexNumList] = leftList[indexLeft]
                #index left moved over one and compares with the old numlist 
                indexLeft += 1
                
                #if on the right side of the list, then the original list index is now index of right list
            else:
                numList[indexNumList] = rightList[indexRight]
                indexRight += 1
                
            indexNumList += 1
            
        #handling leftovers on left side
        while indexLeft < len(leftList):
            numList[indexNumList] = leftList[indexLeft]
            indexLeft += 1
            indexNumList += 1
            count = count +1
            
        #handling leftovers on the right side
        while indexRight < len(rightList):
            numList[indexNumList] = rightList[indexRight]
            indexRight += 1
            indexNumList += 1
            count = count +1
    return count     
    #print("The number of merge comparisons is: ", count)


def timeEfficiency():
    #time.time or time.perf_counter()
    numList = []
    readData(numList)
    startIn = time.time()
    insertionSort(numList)
    endIn = time.time()
    
    
    #print results
    print("Insertion Starts at: ", startIn , "seconds")
    print("Insertion Ends at: ", endIn , "seconds")
    print("Insertion Time taken: ", endIn- startIn , "seconds")
    
    
    
    startMe = time.time()
    mergeSort(numList)
    endMe = time.time()
    
    
    #print results
    countMe = mergeSort(numList)
    print("\n\nThe number of merge comparisons is: ", countMe)
    print("Merge Starts at: ", startMe , "seconds")
    print("Merge Ends at: ", endMe , "seconds")
    print("Merge Time taken: ", endMe- startMe , "seconds")
    
timeEfficiency() 
```

3.1  Write down a general description of the loop invariant technique in your own words as a proof technique of correctness.  (2 points)

A loop invariant is a statement that needs to be satisfied at the beginning of every loop
It also helps you understand why an algorithm gives you the correct answer



3.2 Identify the loop invariant of the loop in your merge() function (3 points)
My loop invariant is to make sure that the current(index) is less than the numberList of the index 
The outer loop: while indexLeft < len(leftList) and indexRight < len(rightList) 
is my loop invariant. It shows the start of the loop and that assumes that the numList[indexnumList] is currently empty or contains the smallest number 


3.3  Describe initialization step (0 points)
Initialization step is to make sure that the loop invariant is satisfied so it can continue looping

3.4  Describe maintenance  step (4 points)
If the loop invariant is true before the ith iteration, then the loop invariant will be true before the i + 1st iteration.
As the sort moves forward, it checks to make sure that the left number is smaller than the right number, then moves onto the next position and compares that one. 
It goes numList[indexNumList] = leftList[indexLeft] where it re-establishes that the numList[indexnumlist] now contains the smallest numbers in the list


3.5  Describe Termination step (1 point)
When the loop terminates, the loop invariant shows that it is true and that helps show that the algorithm is correct
Basically when any of the while loop conditions can no longer be fulfilled, 
while indexLeft < len(leftList) and indexRight < len(rightList) 
is when this can no longer be satisfied and the largest numbers are now added to the list
